### **4. Stopwords**

Stopwords are **common words** in a language that are **filtered out** in Natural Language Processing (NLP) because they **do not carry much meaning**. These include words like *"the," "is," "and," "in," "of,"* etc.  

Since stopwords appear **frequently** in text but add little value to analysis, they are often **removed** to improve efficiency in NLP tasks like text classification, search engines, and machine learning models.

---

**Examples of Stopwords in English** 
 
| Category | Examples |
|----------|----------|
| Articles | the, a, an |
| Pronouns | he, she, it, they |
| Conjunctions | and, or, but, nor |
| Prepositions | in, on, at, with |
| Auxiliary Verbs | is, are, was, were, be |
| Other Common Words | this, that, those, these |

---

**Stopwords in NLTK**

NLTK provides a built-in list of stopwords for multiple languages.  

````python
# imports
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
porter_stemmer = PorterStemmer()
snowball_stemmer = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()
````

Apply Stopwords and filter and then apply stemming or lemmatization.

1) Using Porter Stemmer

````python
for i in range(len(speech_sentences)):
    words = nltk.word_tokenize(speech_sentences[i])
    # words = [word for word in words if word not in eng_stopwords]
    words = [porter_stemmer.stem(word) for word in words if word not in eng_stopwords]
    speech_sentences[i] = ' '.join(words) # join the words to form a sentence
speech_sentences
````

2) Using Snowball Stemmer

````python
# reinitialize the variables
speech_sentences = nltk.sent_tokenize(paragraph)
speech_words = nltk.word_tokenize(paragraph)
````

````python
for i in range(len(speech_sentences)):
    words = nltk.word_tokenize(speech_sentences[i])
    # words = [word for word in words if word not in eng_stopwords]
    words = [snowball_stemmer.stem(word) for word in words if word not in eng_stopwords]
    speech_sentences[i] = ' '.join(words) # join the words to form a sentence
speech_sentences
````

3) Using Lemmatization

````python
# reinitialize the variables
speech_sentences = nltk.sent_tokenize(paragraph)
speech_words = nltk.word_tokenize(paragraph)
````

````python
for i in range(len(speech_sentences)):
    words = nltk.word_tokenize(speech_sentences[i])
    # words = [word for word in words if word not in eng_stopwords]
    words = [lemmatizer.lemmatize(word.lower(), pos='v') for word in words if word not in eng_stopwords]
    speech_sentences[i] = ' '.join(words) # join the words to form a sentence
speech_sentences
````


**Why Remove Stopwords?**

✅ **Reduces dataset size** → Speeds up processing.  
✅ **Improves accuracy** → Less noise in text analysis.  
✅ **Enhances model performance** → Focuses on important words.  

---

**When NOT to Remove Stopwords?**

❌ If stopwords **carry meaning** (e.g., in **sentiment analysis** or **question classification**).  
❌ If you are working with **short texts** (e.g., chatbot responses).  

---

### **5. Parts of Speech Tagging**

**Parts of Speech (POS) tagging** is the process of **assigning word categories** (noun, verb, adjective, etc.) to words in a sentence based on their **context and grammatical role**.

---

**POS Tagging in NLTK**

NLTK provides a POS tagger that assigns **Penn Treebank POS tags** to words.  

**Common POS Tags and Their Meanings**
  

| **Tag**  | **Part of Speech**    | **Example**              |
|---------|------------------|--------------------------|
| **NN**   | Noun (Singular)  | cat, car, tree         |
| **NNS**  | Noun (Plural)    | cats, cars, trees      |
| **NNP**  | Proper Noun      | John, India, Google    |
| **NNPS** | Proper Noun (Plural) | Americans, Europeans |
| **VB**   | Verb (Base Form) | eat, run, play         |
| **VBD**  | Verb (Past Tense) | ate, ran, played       |
| **VBG**  | Verb (Gerund)    | eating, running, playing |
| **VBN**  | Verb (Past Participle) | eaten, driven, slept |
| **VBP**  | Verb (Present)   | eat, run, sleep        |
| **VBZ**  | Verb (3rd Person Singular) | eats, runs, sleeps |
| **JJ**   | Adjective        | beautiful, happy, tall |
| **JJR**  | Comparative Adjective | bigger, faster, better |
| **JJS**  | Superlative Adjective | biggest, fastest, best |
| **RB**   | Adverb           | quickly, silently, well |
| **RBR**  | Comparative Adverb | faster, better |
| **RBS**  | Superlative Adverb | fastest, best |
| **IN**   | Preposition      | in, on, at, with       |
| **DT**   | Determiner       | the, a, an, this       |
| **PRP**  | Personal Pronoun | he, she, it, we       |
| **PRP$** | Possessive Pronoun | his, her, their       |
| **CC**   | Coordinating Conjunction | and, but, or |
| **CD**   | Cardinal Number | one, two, 100 |
| **EX**   | Existential "there" | there is, there was |
| **MD**   | Modal Verb | can, should, must |
| **UH**   | Interjection | oh, wow, hey |
| **.**    | Punctuation | ., ?, ! |

---

**Why Use POS Tagging?**

✅ **Text Processing** → Improves parsing and understanding.  
✅ **Named Entity Recognition (NER)** → Helps in identifying names, places, etc.  
✅ **Lemmatization** → More accurate results when using POS information.  
✅ **Machine Learning Models** → Features for text classification and sentiment analysis.

### **6. Named Entity Recognition**

**Named Entity Recognition (NER)** is an **information extraction** technique that identifies and classifies **named entities** (people, places, organizations, dates, etc.) in a text.  

**Example:**  

📜 **Sentence:**  

*"Elon Musk founded SpaceX in 2002 in the United States."*  

🔍 **NER Output:**  

- **Elon Musk** → **PERSON**  
- **SpaceX** → **ORGANIZATION**  
- **2002** → **DATE**  
- **United States** → **GPE (Geopolitical Entity)**  

---
**Types of Named Entities**

NER typically classifies words into categories like:

| **Entity Type** | **Description**                      | **Examples**                      |
|--------------|----------------------------------|----------------------------------|
| **PERSON**   | Names of people                 | Elon Musk, Barack Obama        |
| **ORG**      | Organizations                   | Google, SpaceX, OpenAI         |
| **GPE**      | Countries, cities, states      | United States, India, London   |
| **DATE**     | Dates, years, time periods      | 2020, January 1, 21st century  |
| **TIME**     | Specific times                  | 5 PM, midnight, 2:30 AM        |
| **MONEY**    | Monetary values                 | $10 million, 500 INR           |
| **PERCENT**  | Percentage values               | 50%, 23.4%                     |
| **LOC**      | Non-GPE locations (mountains, rivers) | Mount Everest, Amazon River |
| **EVENT**    | Named events                    | World War II, Olympics 2024    |
| **FAC**      | Buildings, airports, highways  | Eiffel Tower, Heathrow Airport |

---
**NER in NLTK**

NLTK provides **NER tagging** using a **pre-trained model**.

---
**Why is NER Important?**

✅ **Information Extraction** → Extract useful data from unstructured text.  
✅ **Search Optimization** → Improves search engines and recommendation systems.  
✅ **Question Answering Systems** → Helps AI understand important entities in text.  
✅ **Financial & Healthcare Applications** → Identifies companies, stock prices, diseases, and medical terms.
